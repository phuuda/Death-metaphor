```python
from IPython.display import HTML, display
import pandas as pd

import nltk
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.collocations import *
from nltk.corpus import names 

import unicodedata
import collections

import string

#caring_path = "/Users/Sofia/Desktop/LING_506/project/data/data_version_01/cancer_research_caring_03.csv"
caring_path = "/Users/Sofia/Desktop/LING_506/project/data/data_version_01/cancer_research_living_with_cancer_03.csv"

caring_data = pd.read_csv(caring_path, delimiter="\t")
column_names = ['Thread_name', 'Thread_address', 'Username', "Post_id",
                                  "Date", "Post_text", "Reply_to_id"]
```

```python
# find occurrence of 'fight', 'battle', 'stuggle'

# first remove punctuation, words -> lowercase

punctuation_marks = string.punctuation
print(punctuation_marks)
punctuation_marks = [char for char in punctuation_marks]
punctuation_marks.append("“")
print(punctuation_marks)

text_lines = caring_data.Post_text
all_words_list = []

emojis = ["☺️", "☹️", "❤️", "❤"]

all_words_list = []
text_lines_no_punct = []  # DO NOT RESET !
line_count = 0

for line in text_lines:
    new_line = line
    #print(line_count)
    if isinstance(line, str):
        
        new_line = new_line.lower()  # LOWERCASE
        
        for mark in punctuation_marks:  # REMOVE PUNCTUATION
            if mark in line:
                new_line = new_line.replace(mark, " ")
                
        for emoji in emojis:
            if emoji in line:
                new_line = new_line.replace(emoji, " " + emoji + " ")
                
        new_line = unicodedata.normalize("NFKD", new_line)
                
        while "\n" in new_line:
            new_line = new_line.replace("\n", " ")
            
        while "\t" in new_line:
            new_line = new_line.replace("\t", " ")
            
        while "  " in new_line:
            new_line = new_line.replace("  ", " ")
            
        word_list = new_line.split(" ")
        word_list = [word for word in word_list]
        #word_list = [word for word in word_list if word not in stop_words]  # REMOVE STOP WORDS
                      
    text_lines_no_punct.append(new_line)   
    all_words_list.extend(word_list)
    line_count += 1




word_freq = collections.Counter(all_words_list)
word_freq = collections.OrderedDict(word_freq.most_common())
print(len(text_lines_no_punct))
```

```python
#print(len(word_freq))

# 'fight', 'fought'
# 'battle', 'battl'

fight_count = 0 # total num of all 'fight' met occurrences

for key, val in word_freq.items():
    
    if 'fight' in key:
        fight_count += val
        print(key, val)
        
    elif 'fought' in key:
        fight_count += val
        print(key, val)        
        
print(fight_count)
```
